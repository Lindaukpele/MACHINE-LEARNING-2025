```python
#Basic scatter plot
import matplotlib.pyplot as plt
x1 = 2
y1 = 21
point_1 = [x1, y1]
```

```python
#upper plot made using (x1, y1)
plt.scatter(x1, y1)
plt.show()

#lower plot made using the data extracted from the point
plt.scatter(point_1[0], point_1[1])
plt.show()

#These two plots show the exact same thing because the data is the same 
```

```python
#Value X Corresponds to the x-axis (horizontal) and y value corresponds to y-axis (vertical axis)
#plt.scatter(x1, y1)  OR
plt.scatter(point_1[0], point_1[1])
plt.xlabel("Horizontal Axis - Independent Variable")
plt.ylabel("Vertical Axis - Dependent Variable")
plt.xlim(0,3)   #shows limit of x axis from 0 to 3
plt.ylim(0,30)   #shows limit of y axis from 0 to 30

plt.show()    #Forces the notebook to show the output
         
```

```python
#Lets say we have a data set
#X Measurements:
x1 = 2
x2 = 3
x3 = 4
x4 = 5
x5 = 2.5

#y Measurements:
y1 = 21
y2 = 32
y3 = 39
y4 = 50
y5 = 24

x_values = [x1, x2, x3, x4, x5]
y_values = [y1, y2, y3, y4, y5]

print("x-values = ", x_values, "\ny-values = ", y_values)
```

```python
plt.scatter(x_values, y_values)
plt.xlabel("Horizontal Axis - Independent Variable")
plt.ylabel("Vertical Axis - Dependent Variable")
plt.xlim(0,6), plt.ylim(0,60)

plt.show()    
```

```python
#Thwew is a relationship. When the value of x is increased, so is the value of y. Therefore, we can say that the y value is dependent on the x value. In this case, 
#...because it looks like we could draw a line straight through all points, we could also say that it is highly, linearly positively correlated.
```

```python
plt.scatter(x_values, y_values)
plt.plot(range(1,7,1), range(10,70,10), color = 'red')  # Draws a line to show the relationship. range starts at 1, stop at 7 increasing by 1 i.e 1,7,1
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(0,6), plt.ylim(0,60)

plt.show()    
```

We can also plot negative numbers on a scatter plot

```python
x_values = [-9, -2, 0, 4, 9]
y_values = [8, 4, 0, -5, -8]

plt.axhline(color = 'black'), plt.axvline(color = 'black')
plt.scatter(x_values, y_values, s = 100)
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
```

```python
#Adding the trend line helps us make predictions of where new data could fir in our relationship.

plt.axhline(color = 'black'), plt.axvline(color = 'black')
plt.scatter(x_values, y_values, s = 100)
plt.plot([-9.5,9.5], [9.5,-9.5], color = 'red')     #why did we use square bracket here instead of range as the previous??????????
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
```

```python
plt.axhline(color = 'black'), plt.axvline(color = 'black')
plt.scatter(x_values, y_values, s = 100)
plt.plot([-9.5,9.5], [9.5,-9.5], color = 'red')     #why did we use square bracket here instead of range as the previous??????????
plt.scatter(6, -6, s = 100, color = 'green')
plt.plot([6,6], [-6,-10], color = 'green'), plt.plot([-10,6], [-6,-6], color = 'green')
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
```

We can do one better and actually calculate hte line of best fit instead of eyeballing it.

```python
#fine line of best fit
import numpy as np
x_values = np.array([-9,-2,0,4,9])

m, b = np.polyfit(x_values, y_values,1)
#fits is y = mx + b formula
```

```python
plt.axhline(color = 'black'), plt.axvline(color = 'black')
plt.scatter(x_values, y_values, s = 100)
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.xlim(-10,10), plt.ylim(-10,10)
plt.plot(x_values, m*x_values+b, color = 'red', linestyle = '--', linewidth=2)
```

```python
#To calculate the actual estimate of the new object with measurement x = 6

x_new = 6
y_new = m*x_new+b
print(y_new)  #pretty close to our eyeball estimate!
```
